```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import precision_score, recall_score, confusion_matrix

# Instantiate LDA
lda = LinearDiscriminantAnalysis()

# Fit the model
lda.fit(X_train, y_train)

print(lda.intercept_, lda.coef_)

# Make predictions
pred_y = lda.predict(X_test)

# Calculate precision, recall, and confusion matrix
precision = precision_score(y_test, pred_y)
recall = recall_score(y_test, pred_y)
conf_matrix = confusion_matrix(y_test, pred_y)

# Display results
precision, recall, conf_matrix
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier

sc_tr =  StandardScaler().fit(X_train)
X_train = sc_tr.transform(X_train)
sc_te = StandardScaler().fit(X_test)
X_test = sc_te.transform(X_test)

params = {'n_neighbors' : range(1,11,1)}
gs = GridSearchCV(KNeighborsClassifier(),params)

gs.fit(X_train,y_train)
gs.best_params_
```

```python
best_knn = gs.best_estimator_

y_pred = best_knn.predict(X_test)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred)

precision, recall, conf_matrix, roc_auc
```
